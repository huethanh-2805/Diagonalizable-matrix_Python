- **B∆∞·ªõc 2:** T√¨m t·∫•t c·∫£ c√°c nghi·ªám $Œª_i$ c·ªßa ph∆∞∆°ng tr√¨nh ƒëa th·ª©c ƒë·∫∑c tr∆∞ng. V·ªõi m·ªói tr·ªã ri√™ng $Œª_i$ t√¨m c∆° s·ªü v√† s·ªë chi·ªÅu cho kh√¥ng gian nghi·ªám ph∆∞∆°ng tr√¨nh $(A ‚àí ŒªI_n) X = 0$.
<br>N·∫øu m·ªói $Œª_i$ s·ªë chi·ªÅu kh√¥ng gian nghi·ªám nh·ªè h∆°n l≈©y th·ª´a c·ªßa $Œª_i$ trong ƒëa th·ª©c ƒë·∫∑c tr∆∞ng th√¨ *A* kh√¥ng ch√©o h√≥a ƒë∆∞·ª£c v√† thu·∫≠t to√°n k·∫øt th√∫c, ng∆∞·ª£c l·∫°i chuy·ªÉn sang ***b∆∞·ªõc 3***.
- **B∆∞·ªõc 3:** V·ªõi c√°c vector trong c∆° s·ªü kh√¥ng gian nghi·ªám t√¨m ƒë∆∞·ª£c ·ªü ***b∆∞·ªõc 2***, ta ƒë·∫∑t ma tr·∫≠n *P* l√† ma tr·∫≠n c√≥ ƒë∆∞·ª£c b·∫±ng c√°ch d·ª±ng c√°c vector th√†nh c√°c c·ªôt. Khi ƒë√≥ ma tr·∫≠n *P* l√†m ch√©o *A* v√† $P^{‚àí1}AP$ l√† ma tr·∫≠n ƒë∆∞·ªùng ch√©o. $diag(Œª_1, . . . , Œª_r)$

# 1) Ch∆∞∆°ng tr√¨nh t√¨m ma tr·∫≠n ch√©o  $ùëÉ,ùëÉ^{‚àí1}$ v√† ma tr·∫≠n ƒë∆∞·ªùng ch√©o D

## C√°c h√†m x·ª≠ l√Ω ph·ª•

```python
# H√†m nh√¢n ma tr·∫≠n v·ªõi m·ªôt h·∫±ng s·ªë alpha
def multiplyMatrix_Const(A, alpha):
    B = [[A[i][j]*alpha for j in range(len(A[0]))] for i in range(len(A))]
    return B

#Tr·ª´ hai ma tr·∫≠n
def sub_Matrix(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        return None
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

#H√†m nh√¢n hai ma tr·∫≠n
def mul_Matrix(A, B):
    if len(A[0]) != len(B):
        return None
    result = [[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(A[0])):
                result[i][j] = check_number(result[i][j] + A[i][k] * B[k][j])  
    return result

#H√†m l·∫•y ra m·ªôt c·ªôt th·ª© i c·ªßa ma tr·∫≠n
def getColumn(A, i):
    col = []
    for j in range(len(A)):
        col.append(A[j][i])
    return col

#H√†m nh√¢n vector v·ªõi h·∫±ng s·ªë
def mul_Vector_Alpha(A, const):
    return [A[i]*const for i in range(len(A))]   

#H√†m t·∫°o ma tr·∫≠n ƒë∆∞·ªùng ch√©o t·ª´ c√°c gi√° tr·ªã tr√™n ƒë∆∞·ªùng ch√©o cho tr∆∞·ªõc
def diagonal_Matrix(values):
    n = len(values)
    return [[0 if i != j else values[i] for j in range(n)] for i in range(n)]

#H√†m t·∫°o ma tr·∫≠n ƒë∆°n v·ªã
def unit_Matrix(n):
    return [[1 if i == j else 0 for j in range(n)] for i in range(n)]  

def Gauss_elimination(A):
    rows, cols = len(A), len(A[0]) #L·∫•y s·ªë d√≤ng, s·ªë c·ªôt ma tr·∫≠n A
    r = -1 #L∆∞u gi·ªØ d√≤ng ƒë√£ th·ª±c hi·ªán
    for j in range(cols - 1):
        t, r = 0, r + 1 ##Co sua
        if r == rows: break
        for i in range(r, rows):  #X√°c ƒë·ªãnh c·ªôt tr√°i nh·∫•t kh√¥ng ch·ª©a ho√†n to√†n s·ªë 0    
            if A[i][j] == 0:
                t = t + 1
                if t == (rows - r):
                    j = j + 1
                    i = r      
            if A[i][j] != 0:
                c = j
                break        
        if A[r][c] == 0:  #ƒê·ªïi ch·ªó 2 d√≤ng n·∫øu c·∫ßn
            for k in range(r, rows):
                if A[k][c] != 0:
                    A[k], A[r] = A[r], A[k]
                    break
        for l in range(c, cols): #Nh√¢n v·ªõi 1/a 
            if A[r][l] != 0:
                a = A[r][l]
                break
        for m in range(c, cols):
            if A[r][m] == 0:
                A[r][m] = A[r][m] / 1
                continue
            A[r][m] = A[r][m] / a
        for i in range(r + 1, rows): #Tr·ª´ c√°c d√≤ng 
            if A[i][c] != 0:
                x = A[i][c]
                for n in range(c, cols):
                    A[i][n] = A[i][n] - A[r][n]*x              
    return A 

#H√†m c·ªông 2 vector
def add_Vector(vector1, vector2):
    return [vector1[i] + vector2[i] for i in range(len(vector1))]   

#H√†m cho 2 vector c·ªôt b·∫±ng nhau
def equal_Col(B, k, vector):
    for i in range(len(B)):
        B[i][k] = vector[i]    

#H√†m ki·ªÉm tra d√≤ng to√†n 0
def check_row_iszero(B, i):
    for j in range(len(B)):
        if B[i][j] != 0:
            return False
    return True    

#H√†m t·∫°o ma tr·∫≠n b·ªï sung
def create_Additional_Matrix(A):
    n = len(A)
    result = [[0 if i == n else A[j][i] for i in range(n+1)] for j in range(n)]
    return result 

#H√†m l∆∞u vector ri√™ng
def private_Vector(R, check, index):
    for i in range(len(R)):
        if (check_row_iszero(R, i) == False):
            index += 1
            for j in range(len(R)):
                check[j][index] = R[i][j]   
    return index            
        
def back_substitution(A):
    #Bi·ªán lu·∫≠n d√≤ng to√†n s·ªë 0:
    rows, cols = len(A), len(A[0])
    i = rows - 1
    while i >= 0:
        count = 0 #ƒê·∫øm s·ªë s·ªë 0 tr√™n d√≤ng i
        for j in range(cols):
            if A[i][j] == 0:
                count = count + 1
            if count == (cols - 1) and A[i][cols - 1] != 0: #Khi d√≤ng g·ªìm to√†n s·ªë 0 c·ªßa A m√† h·ªá s·ªë v·∫ø ph·∫£i (t·ª± do) l√† s·ªë # 0 th√¨ k·∫øt lu·∫≠n h·ªá v√¥ nghi·ªám
                return None #print("H·ªá ph∆∞∆°ng tr√¨nh v√¥ nghi·ªám") 
            if count == cols:
                rows = rows - 1 #b·ªè d√≤ng g·ªìm to√†n s·ªë 0
        i = i - 1
    #Ph√©p th·∫ø ng∆∞·ª£c
    #V√¥ s·ªë nghi·ªám: xu·∫•t ra t·ª´ng nghi·ªám ·ªü d·∫°ng chu·ªói
    if cols - 1 > rows: #"H·ªá v√¥ s·ªë nghi·ªám"
        B = [[0 for j in range(len(A))] for i in range(len(A))] #L∆∞u nghi·ªám x1,x2,x3,... c·ªßa h·ªá
        i = rows - 1 
        trace = [0]  * (cols - 1) #ƒê√°nh d·∫•u ƒë√£ ƒë·∫∑t ·∫©n hay ch∆∞a
        while i >= 0: 
            dem = 0 #ƒê·∫øm s·ªë 0 tr√™n d√≤ng i t·ª´ tr√°i->ph·∫£i cho ƒë·∫øn khi g·∫∑p h·ªá s·ªë # 0 ƒë·∫ßu ti√™n th√¨ d·ª´ng l·∫°i
            for j in range(cols - 1):
                if A[i][j] == 0:
                    dem = dem + 1
                if A[i][j] != 0:
                    break    
            k = cols - 2  #ƒê·∫∑t ·∫©n
            t0 = 0
            while k >= dem:
                if k != dem:
                    if trace[k] == 1:
                        k = k - 1
                        continue
                    else:
                        B[t0][k] = 1
                        trace[k] = 1 
                        t0 = t0 + 1
                else:
                    vector = [0] * len(B)
                    for m in range(k+1, cols-1):
                        col_B = getColumn(B, m)
                        vector = add_Vector(vector, mul_Vector_Alpha(col_B, -A[i][m])) 
                    equal_Col(B, k, vector) 
                    trace[k] = 1 #L∆∞u v·∫øt ƒë√£ t√≠nh xong x th·ª© k
                k = k - 1             
            i = i - 1
        return B
    #"H·ªá c√≥ nghi·ªám duy nh·∫•t"
    B = [0] * (cols - 1)
    i = rows - 1
    while i >= 0:
        s = 0
        j = cols - 1
        if A[i][i] == 1:
            for k in range(cols - 1):
                s = float("{:.2f}".format(s + A[i][k] * B[k]))   
            B[i] = float("{:.2f}".format(A[i][cols - 1] - s))
        i = i - 1   
    return B 

#H√†m gi·∫£i HPT (A-ŒªI_n)X = 0
def solve_Equations(A, lamda, n):
    B = sub_Matrix(A, multiplyMatrix_Const(unit_Matrix(n), lamda))
    R1 = Gauss_elimination(B) #Ph√©p kh·ª≠ Gauss
    R = create_Additional_Matrix(R1)
    return back_substitution(R) #Ph√©p th·∫ø ng∆∞·ª£c

#H√†m ki·ªÉm tra sai s·ªë
import math
def check_number(number):
    if math.isclose(number,math.ceil(number) ,abs_tol=1e-20):
        return math.ceil(number)
    elif math.isclose(number,math.floor(number) ,abs_tol=1e-20):
        return math.floor(number)
    elif math.isclose(number,round(number,2) ,abs_tol=1e-20):
        return round(number,2)
    else:
        return number

#H√†m t√¨m ma tr·∫≠n ngh·ªãch ƒë·∫£o
def inverse(A):
    n = len(A) #A l√† ma tr·∫≠n vu√¥ng c·∫•p n
    I = [[1 if i == j else 0 for j in range(n)] for i in range(n)]     #T·∫°o ma tr·∫≠n ƒë∆°n v·ªã 
    #Gi·∫£i thu·∫≠t v·ªõi ph√©p kh·ª≠ Gauss ‚Äì Jordan
    for i in range(n):
        t = 0  #Bi·∫øn ƒë·∫øm s·ªë s·ªë 0 t·ª´ d√≤ng i ƒë·∫øn d√≤ng n tr√™n c·ªôt i
        for j in range(i, n):  #B∆∞·ªõc 1:
            if A[j][i] == 0:
                t = t + 1       
        if t == (n - i): #N·∫øu ch√∫ng g·ªìm to√†n s·ªë 0
            print("Ma tr·∫≠n kh√¥ng kh·∫£ ngh·ªãch")
            return 0
        if A[i][i] == 0: #Ng∆∞·ª£c l·∫°i ƒë·ªïi ch·ªó 2 d√≤ng n·∫øu c·∫ßn thi·∫øt
            for k in range(i, n):
                if A[k][i] != 0:
                    A[k], A[i] = A[i], A[k]
                    I[k], I[i] = I[i], I[k]
                    break     
        if A[i][i] != 0: #B∆∞·ªõc 2: Nh√¢n v·ªõi 1/a       
            a = A[i][i]
        for m in range(n): 
            A[i][m] = A[i][m] / a 
            I[i][m] = I[i][m] / a
        for j in range(n): #B∆∞·ªõc 3: C·ªông m·ªôt b·ªôi s·ªë th√≠ch h·ª£p c·ªßa d√≤ng i v·ªõi c√°c d√≤ng kh√°c d√≤ng i              
            if j == i:     # ƒë·ªÉ bi·∫øn c√°c s·ªë h·∫°ng tr√™n c·ªôt i v·ªÅ s·ªë 0 (tr·ª´ s·ªë h·∫°ng n·∫±m ·ªü d√≤ng i)
                continue
            if A[j][i] != 0:
                x = A[j][i]
                for k in range(n):
                    A[j][k] = A[j][k] - A[i][k]*x 
                    I[j][k] = I[j][k] - I[i][k]*x                      
    for i in range(n):
        for j in range(n):
            I[i][j] = check_number(I[i][j])
    
    return I      

#H√†m ki·ªÉm tra xem n·∫øu c√≥ gi√° tr·ªã lamda l·∫∑p l·∫°i th√¨ ch·ªâ gi·ªØ l·∫°i 1 lamda
def check_different(vector): 
    for i in range(len(vector)):
        for j in range(i+1, len(vector)):
            if check_number(vector[i]) == check_number(vector[j]):
                vector[j] = -99999
           
```

## H√†m t√¨m ma tr·∫≠n ch√©o  $ùëÉ,ùëÉ^{‚àí1}$ v√† ma tr·∫≠n ƒë∆∞·ªùng ch√©o $D$

```python
import numpy as np

def diagonalizable_Matrix(A):
    n = len(A)
    w,v = np.linalg.eig(A)
    check_different(w)
    #Gi·∫£i t·ª´ng tr·ªã ri√™ng
    check = [[0 for i in range(n)] for j in range(n)]
    index = -1
    for i in range(len(w)):
        lamda = round(w[i],2)
        if lamda != -99999:
            R = solve_Equations(A, lamda, n)
            index = private_Vector(R, check, index)
    if index == n - 1:
            P = check
            P_inverse = inverse(P) #T√≠nh P^-1
            D0 = mul_Matrix(P_inverse, A)
            P = inverse(P_inverse)  
            D = mul_Matrix(D0, P)
    else:
        print("Ma tr·∫≠n kh√¥ng ch√©o h√≥a ƒë∆∞·ª£c v√¨ ch·ªâ c√≥", index+1, "vector ri√™ng")
        return None
    return P, D

A = [[3,4,-4],[-2,-1,2],[-2,0,1]]

if diagonalizable_Matrix(A) != None:
    P, D = diagonalizable_Matrix(A)
    print("- Ma tr·∫≠n ch√©o P:", P)
    print("- Suy ra P^(-1):", inverse(P), "\n- Ma tr·∫≠n ƒë∆∞·ªùng ch√©o D:", D)
```

## So s√°nh k·∫øt qu·∫£:
Khi s·ª≠ d·ª•ng th∆∞ vi·ªán **"sympy"** v√† d√πng h√†m **"Matrix.diagonalize()"** ƒë·ªÉ t√¨m ma tr·∫≠n ch√©o  $ùëÉ,ùëÉ^{‚àí1}$ v√† ma tr·∫≠n ƒë∆∞·ªùng ch√©o $ùê∑$ cho ra k·∫øt qu·∫£ c√°c ma tr·∫≠n c√≥ s·ª± kh√°c nhau v·ªÅ th·ª© t·ª± s·∫Øp x·∫øp c√°c vector ri√™ng(t√πy thu·ªôc v√†o tr·ªã ri√™ng n√†o t√≠nh tr∆∞·ªõc). Nh∆∞ng v·∫´n ƒë·∫£m b·∫£o $A = PDP^{-1}$

## ·ª®ng d·ª•ng c·ªßa ch√©o h√≥a

### 1. T√≠nh to√°n hi·ªáu qu·∫£ l≈©y th·ª´a c·ªßa m·ªôt ma tr·∫≠n
Cho ma tr·∫≠n vu√¥ng A, gi·∫£ s·ª≠ A ch√©o h√≥a ƒë∆∞·ª£c.

Khi ƒë√≥, ta c√≥:

$P^{-1}AP = D = \begin{pmatrix} \lambda_1 & 0 & ... & 0 \\ 
                     0 & \lambda_2 & ... & 0\\ 
                     \vdots &\vdots  & \ddots &\vdots\\                  
                     0 & 0 & ... & \lambda_n \end{pmatrix}$
                     $<=> A = PDP^{-1} <=> A^k = PD^kP^{-1} 
<=> A^k = P\begin{pmatrix} \lambda_1^k & 0 & ... & 0 \\ 
                     0 & \lambda_2^k & ... & 0\\ 
                     \vdots &\vdots  & \ddots &\vdots\\                  
                     0 & 0 & ... & \lambda_n^k \end{pmatrix}P^{-1}$
                     
***V√≠ d·ª•:*** $A=\begin{pmatrix} 3 & 4 & -4 \\ -2 & -1 & 2\\ -2 & 0 & 1 \end{pmatrix}$. T√≠nh $A^{100}$      

Sau khi ch√©o h√≥a ma tr·∫≠n t√≠nh ƒë∆∞·ª£c: $P=\begin{pmatrix} 1 & 0 & -1 \\ 0 & 1 & 1\\ 1 & 1 & 1 \end{pmatrix}$ => $P^{-1}=\begin{pmatrix} 0 & -1 & 1 \\ 1 & 2 & -1\\ -1 & -1 & 1 \end{pmatrix}$ v√† $P^{-1}AP=D=\begin{pmatrix} -1 & 0 & 0 \\ 0 & 1 & 0\\ 0 & 0 & 3 \end{pmatrix}$

Khi ƒë√≥, ta c√≥:

$A^{100}=PD^{100}P^{-1}=\begin{pmatrix} 1 & 0 & -1 \\ 0 & 1 & 1\\ 1 & 1 & 1 \end{pmatrix}\begin{pmatrix} (-1)^{100} & 0 & 0 \\ 0 & 1^{100} & 0\\ 0 & 0 & 3^{100} \end{pmatrix}\begin{pmatrix} 0 & -1 & 1 \\ 1 & 2 & -1\\ -1 & -1 & 1 \end{pmatrix}$

### 2. T√¨m s·ªë h·∫°ng t·ªïng qu√°t c·ªßa d√£y s·ªë

***√ù t∆∞·ªüng:*** T·ª´ c√¥ng th·ª©c truy h·ªìi c·ªßa d√£y s·ªë ${U_n}$. X√°c ƒë·ªãnh ma tr·∫≠n A v√† ma tr·∫≠n ${X_n}$ sao cho c√≥ ƒë∆∞·ª£c c√¥ng th·ª©c
$X_n= AX_{n-1}$. T·ª´ ƒë√≥ suy ra $X_n=A^nX_0$.
- Ch√©o h√≥a, tam gi√°c h√≥a ma tr·∫≠n A.
- Suy ra $X_n = PD^nP^{-1}X_0$ or $X_n = PT^nP^{-1}X_0$. K·∫øt lu·∫≠n ƒë∆∞·ª£c $U_n$

***V√≠ d·ª•***: Cho $U_0=1, U_1=1$ v√† $U_n = U_{n-1}+6U_{n-2},$ v·ªõi m·ªçi n >= 2

ƒê·∫∑t $A = \begin{bmatrix} 0 & 1\\ 6 & 1 \end{bmatrix}, X_n=\begin{bmatrix} u_n \\ u_{n+1} \end{bmatrix}$ => $X_n= AX_{n-1}=A^nX_0$

Ph∆∞∆°ng tr√¨nh ƒë·∫∑c tr∆∞ng c·ªßa A: $det(A-\lambda I)=0$ <=> $\lambda^2 - \lambda-6=0$ <=> $\lambda = 3 ho·∫∑c \lambda = -2$
